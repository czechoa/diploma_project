# Uczenie na zbiorze polemo2 a ocena na zbiorze Alegro

```python
polemo2 = db_dict['clarin-pl/polemo2-official'].data
train = polemo2['train'].to_pandas()
validation = polemo2['validation'].to_pandas()
test = polemo2['test'].to_pandas()

# 1 = negatuwne
# 2 = pozytywbe
# 0 = neurealne
# 3 = dwuczanczne
binary_train = train[(train['target'] == 1) | (train['target'] == 2)]
binary_val = validation[(validation['target'] == 1) | (validation['target'] == 2)]
binary_test = test[(test['target'] == 1) | (test['target'] == 2)]

from machine_learning.utils.utils import get_unicode_text

token_pattern = " "

text_train, text_val, text_test = get_unicode_text('token_tekst', binary_train, binary_val, binary_test, token_pattern)

binary_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.pipeline import Pipeline
import numpy as np
```

```python

text_clf = Pipeline(
    [
        ("vect", CountVectorizer(strip_accents='ascii')),
        ("tfidf", TfidfTransformer()),
        ("clf", LogisticRegression(max_iter=1000, n_jobs=-1, class_weight='balanced')),
    ]
)
parameters = {
    'vect__ngram_range': [(1, 3)],
    'vect__min_df': [2,3,4],
    'tfidf__use_idf': (True, False),
    'clf__C': ( 15,20,30)
}
text_grid = np.hstack([text_train.values, text_val.values])
target_grid =  np.hstack([binary_train['target'].values ,binary_val['target'].values])

gs_clf = GridSearchCV(text_clf, parameters, cv=5, n_jobs=-1)

gs_clf = gs_clf.fit(text_grid, target_grid)

print("the best mean score", gs_clf.best_score_, "std test score",
      gs_clf.cv_results_['std_test_score'][gs_clf.best_index_])

for param_name in sorted(parameters.keys()):
    print("%s: %r" % (param_name, gs_clf.best_params_[param_name]))
```

```python
text_clf = Pipeline(
    [
        ("vect", CountVectorizer(strip_accents='ascii',ngram_range =(1, 3), min_df=3)),
        ("tfidf", TfidfTransformer()),
        ("clf", LogisticRegression(max_iter=1000, n_jobs=-1,C=20,class_weight='balanced')),
    ]
)


text_clf.fit(text_train, binary_train['target'].values)
text_clf.score(text_val,binary_val['target'].values)
```

```python
from unidecode import unidecode

ambiguous = train[train['target'] ==3 ]
text_ambiguous = ambiguous['token_tekst'].str.join(token_pattern).apply(unidecode)
predict_proba = text_clf.predict_proba(text_ambiguous)
predict_ambiguous = pd.DataFrame({'text': ambiguous['text_org'],'negative': predict_proba[:, 0],'positive': predict_proba[:,1]})
print('najbardziej pozytywna \n')

print(predict_ambiguous[predict_ambiguous['positive'] == predict_ambiguous['positive'].max()]['text'].values)
print()
print('najbardziej negatywna \n')
print(predict_ambiguous[predict_ambiguous['negative'] == predict_ambiguous['negative'].max()]['text'].values)
predict_ambiguous.to_csv("predict_ambiguous.csv",index=False)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.pipeline import Pipeline
import numpy as np

text_clf = Pipeline(
    [
        ("vect", CountVectorizer(strip_accents='ascii')),
        ("tfidf", TfidfTransformer()),
        ("clf", LogisticRegression(max_iter=1000, n_jobs=-1)),
    ]
)
parameters = {
    'vect__ngram_range': [(1,4),(1, 5)],
    'vect__min_df': [2,4,6],
    # 'tfidf__use_idf': (True, False),
    'clf__C': (10,20,30)
}
text_grid = np.hstack([text_train.values, text_val.values])
target_grid =  np.hstack([binary_train['rating'].values ,binary_val['rating'].values])

gs_clf = GridSearchCV(text_clf, parameters, cv=5, n_jobs=-1)

gs_clf = gs_clf.fit(text_grid, target_grid)

print("the best mean score", gs_clf.best_score_, "std test score",
      gs_clf.cv_results_['std_test_score'][gs_clf.best_index_])

for param_name in sorted(parameters.keys()):
    print("%s: %r" % (param_name, gs_clf.best_params_[param_name]))
```
